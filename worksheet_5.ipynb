```python
import numpy as np


def gradient_descent(X, Y, W, alpha, iterations):
    """
    Perform gradient descent to optimize the parameters of a linear regression model.

    Parameters:
    X (numpy.ndarray): Feature matrix (m x n)
    Y (numpy.ndarray): Target vector (m,)
    W (numpy.ndarray): Initial guess for parameters (n,)
    alpha (float): Learning rate
    iterations (int): Number of iterations

    Returns:
    W_update (numpy.ndarray): Optimized weights
    cost_history (list): Cost value at each iteration
    """

    cost_history = [0] * iterations
    m = len(Y)

    for iteration in range(iterations):

        Y_pred = np.dot(X, W)

        loss = Y_pred - Y

        dw = (1 / m) * np.dot(X.T, loss)

        W = W - alpha * dw

        cost = cost_function(X, Y, W)
        cost_history[iteration] = cost

    return W, cost_history
np.random.seed(0)

X = np.random.rand(100, 3)
Y = np.random.rand(100)
W = np.random.rand(3)

alpha = 0.01
iterations = 1000

final_params, cost_history = gradient_descent(X, Y, W, alpha, iterations)

print("Final Parameters:", final_params)
print("Cost History (first 10 values):", cost_history[:10])
print("Final Cost:", cost_history[-1])
```

```python
# ================================
# Step 4: Model Evaluation
# RMSE and R-Squared from Scratch
# ================================

import numpy as np

# -------------------------------
# RMSE Function
# -------------------------------
def rmse(Y, Y_pred):
    """
    This Function calculates the Root Mean Square Error.

    Input Arguments:
    Y: Array of actual (Target) dependent variables.
    Y_pred: Array of predicted dependent variables.

    Output:
    rmse: Root Mean Square Error
    """
    n = len(Y)
    rmse = np.sqrt(np.sum((Y - Y_pred) ** 2) / n)
    return rmse


# -------------------------------
# R-Squared Function
# -------------------------------
def r2(Y, Y_pred):
    """
    This Function calculates the R Squared value.

    Input Arguments:
    Y: Array of actual (Target) dependent variables.
    Y_pred: Array of predicted dependent variables.

    Output:
    r2: Coefficient of Determination
    """
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)     # Total Sum of Squares (SST)
    ss_res = np.sum((Y - Y_pred) ** 2)     # Sum of Squared Residuals (SSR)
    r2 = 1 - (ss_res / ss_tot)
    return r2


# -------------------------------
# Example Test Data
# -------------------------------
Y = np.array([3, 5, 7, 9, 11])        # Actual values
Y_pred = np.array([2.8, 5.1, 6.9, 9.2, 10.8])  # Predicted values


# -------------------------------
# Model Evaluation
# -------------------------------
rmse_value = rmse(Y, Y_pred)
r2_value = r2(Y, Y_pred)

print("RMSE:", rmse_value)
print("R-Squared:", r2_value)

```
